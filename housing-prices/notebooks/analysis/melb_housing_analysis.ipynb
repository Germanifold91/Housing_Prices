## 4.1. Data Transformation Function

```python
def data_cleaning(model_df : pd.DataFrame,
                  parameters : List[Any]) -> pd.DataFrame:
    """ 
    """
    # Conversion of selected set of columns to Int64
    columns_to_int = parameters['columns_to_int']
    for column in columns_to_int:
        model_df[column] = model_df[column].astype('Int64')
    
    # Conversion of columns to CAPS
    model_df.columns = [col.upper() for col in model_df.columns]

    return model_df
```

# 5. Data Analysis

## 5.1. Price Distribution

```python
# Plot histogram with matplotlib
plt.hist(melb_housing['Price'], bins=20)  # Adjust the bins as necessary
plt.xlabel('Rounded Number')
plt.ylabel('Frequency')
plt.title('Histogram of Rounded Numbers')
plt.show()
```

```python
# Define bin edges
price_bins = [0, 500000, 1000000, 1500000, 2000000, float('inf')]
# Define bin labels
price_labels = [' 500k', '> 500k to  1M', '> 1M to  1.5M', '> 1.5M to  2M', '> 2M']
# Use pd.cut to categorize the Price into bins
melb_housing['Price Category'] = pd.cut(melb_housing['Price'], bins=price_bins, labels=price_labels, right=True)
```

### Close to 70% of all homes are valued somewhere between 500k and 1.5M as can be seen in the following distribution

```python
# Count the frequency of each category
price_category_counts = melb_housing['Price Category'].value_counts(normalize = True).sort_index()
# Plot bar chart using matplotlib
plt.bar(price_category_counts.index, price_category_counts.values)
plt.xlabel('Price Category')
plt.ylabel('Frequency')
plt.title('Distribution of Price Categories')
plt.xticks(rotation=45)  # Rotate labels to make them readable
plt.show()
```

## 5.2. Age Distribution

```python
# Define bin edges
age_bins = [-float('inf'), 10, 20, 50, 100, float('inf')]
# Define bin labels
age_labels = [' 10 years', '> 10y to  20y', '> 20y to  50y', '> 50y to  100y', '> 100y']
# Use pd.cut to categorize the age_building into bins
melb_housing['Age Category'] = pd.cut(melb_housing['age_building'], bins=age_bins, labels=age_labels, right=True)
```

### As seens in the following chart most homes are between 50 to 100 years old accounting for nearly 60% of them

```python
# Calculate proportions
age_category_proportions = melb_housing['Age Category'].value_counts(normalize=True).sort_index()
# Plot bar chart of proportions
age_category_proportions.plot(kind='bar')
plt.xlabel('Age Category')
plt.ylabel('Proportion')
plt.title('Proportion of Age Categories')
plt.xticks(rotation=45)  # Rotate labels to make them readable
plt.show()
```

### 5.2.1. How is the distribution of prices in the context of age categories?

It seems that older houses are more expensive that their younger counterparts, very few of these buildings are being sold for less than 500k and nearly one in five is being valued at more than 2 million dollars

```python
# Calculate the cross-tabulation of Age Category vs. Price Category with normalized proportions
age_price_matrix = pd.crosstab(index=melb_housing['Age Category'], 
                               columns=melb_housing['Price Category'], 
                               normalize='index')
# Plot stacked bar chart
age_price_matrix.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.xlabel('Age Category')
plt.ylabel('Proportion')
plt.title('Proportion of Price Categories within Each Age Category')
plt.xticks(rotation=45)
plt.legend(title='Price Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
